### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a figure with a 2x2 layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot in the first subplot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")

# Bar chart in the second subplot
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot in the third subplot
axes[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
for i in range(len(scatter_x)):
    axes[1, 0].text(scatter_x[i], scatter_y[i], f'({scatter_x[i]}, {scatter_y[i]})', fontsize=12, ha='right', va='bottom')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")
axes[1, 0].grid(True)

# Pie chart in the fourth subplot
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
