### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)
y = x ** 2  # Square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Customize the plot
plt.title("Line Plot of x vs x²")
plt.xlabel("x values")
plt.ylabel("y values (x²)")
plt.grid(True)

# Show the plot
plt.show()
```
